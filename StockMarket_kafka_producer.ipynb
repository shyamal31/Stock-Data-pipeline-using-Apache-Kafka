```python
import pandas as pd
from kafka import KafkaProducer
from time import sleep
import json
```

```python
producer = KafkaProducer(bootstrap_servers = ['3.92.217.233:9092'], value_serializer = lambda x: json.dumps(x).encode('utf-8'))
```

```python
df = pd.read_csv('/content/MSFT.csv')
```

```python
while True:
  dict_msft_stock = df.sample(1).to_dict(orient = 'records')[0]
  producer.send('demo_test',dict_msft_stock)
  sleep(10)
```
